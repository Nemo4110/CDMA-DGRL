```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

path_dfs = r"../results/dfs_new"
list_csvgz4drugs = [f for f in os.listdir(path_dfs) if f[-3:]=="csv"  and f[:6]=="4DRUGS"]
list_csvgz4items = [f for f in os.listdir(path_dfs) if f[-3:]=="csv"  and f[:6]=="4LABIT"]
list_csvgz4drugs, list_csvgz4items
```

```python
dfs4drugs = [pd.read_csv(os.path.join(path_dfs, f)) for f in list_csvgz4drugs]
dfs4drugs[0]
```

```python
dfs4items = [pd.read_csv(os.path.join(path_dfs, f)) for f in list_csvgz4items]
dfs4items[0]
```

## Visualization

```python
fmts = ('.C7:', '.C5--', '.C2:', '.C3-')
timesteps = range(1, 20+1)

fig, axs = plt.subplots(7, 2, sharex="all", sharey="row", figsize=(3.5*2, 2.5*7))

# AUC
axs[0, 0].set_title('Medications')
axs[0, 0].plot(timesteps, total_df_diff_gnn4drugs.loc['GAT'].auc, fmts[1],
               timesteps, total_df_diff_gnn4drugs.loc['GEN'].auc, fmts[3],
               timesteps, total_df_diff_gnn4drugs.loc['GIN'].auc, fmts[0])
axs[0, 0].set_ylabel('ROCAUC')
axs[0, 0].legend(["GAT", "GEN", "GIN"])
axs[0, 0].grid(True, linestyle='--', linewidth=.57)
axs[0, 1].set_title('Lab tests')
axs[0, 1].plot(timesteps, total_df_diff_gnn4items.loc['GAT'].auc, fmts[1],
               timesteps, total_df_diff_gnn4items.loc['GEN'].auc, fmts[3],
               timesteps, total_df_diff_gnn4items.loc['GIN'].auc, fmts[0])
axs[0, 1].legend(["GAT", "GEN", "GIN"])
axs[0, 1].grid(True, linestyle='--', linewidth=.57)

# ACC
axs[1, 0].plot(timesteps, total_df_diff_gnn4drugs.loc['GAT'].acc, fmts[1],
               timesteps, total_df_diff_gnn4drugs.loc['GEN'].acc, fmts[3],
               timesteps, total_df_diff_gnn4drugs.loc['GIN'].acc, fmts[0])
axs[1, 0].set_ylabel('Accuracy')
axs[1, 0].legend(["GAT", "GEN", "GIN"])
axs[1, 0].grid(True, linestyle='--', linewidth=.57)
axs[1, 1].plot(timesteps, total_df_diff_gnn4items.loc['GAT'].acc, fmts[1],
               timesteps, total_df_diff_gnn4items.loc['GEN'].acc, fmts[3],
               timesteps, total_df_diff_gnn4items.loc['GIN'].acc, fmts[0])
axs[1, 1].legend(["GAT", "GEN", "GIN"])
axs[1, 1].grid(True, linestyle='--', linewidth=.57)

# JACCARD
axs[2, 0].plot(timesteps, total_df_diff_gnn4drugs.loc['GAT'].jaccard, fmts[1],
               timesteps, total_df_diff_gnn4drugs.loc['GEN'].jaccard, fmts[3],
               timesteps, total_df_diff_gnn4drugs.loc['GIN'].jaccard, fmts[0])
axs[2, 0].set_ylabel('Jaccard')
axs[2, 0].legend(["GAT", "GEN", "GIN"])
axs[2, 0].grid(True, linestyle='--', linewidth=.57)
axs[2, 1].plot(timesteps, total_df_diff_gnn4items.loc['GAT'].jaccard, fmts[1],
               timesteps, total_df_diff_gnn4items.loc['GEN'].jaccard, fmts[3],
               timesteps, total_df_diff_gnn4items.loc['GIN'].jaccard, fmts[0])
axs[2, 1].legend(["GAT", "GEN", "GIN"])
axs[2, 1].grid(True, linestyle='--', linewidth=.57)

# precision
axs[3, 0].plot(timesteps, total_df_diff_gnn4drugs.loc['GAT'].precision, fmts[1],
               timesteps, total_df_diff_gnn4drugs.loc['GEN'].precision, fmts[3],
               timesteps, total_df_diff_gnn4drugs.loc['GIN'].precision, fmts[0])
axs[3, 0].set_ylabel('Precision')
axs[3, 0].legend(["GAT", "GEN", "GIN"])
axs[3, 0].grid(True, linestyle='--', linewidth=.57)
axs[3, 1].plot(timesteps, total_df_diff_gnn4items.loc['GAT'].precision, fmts[1],
               timesteps, total_df_diff_gnn4items.loc['GEN'].precision, fmts[3],
               timesteps, total_df_diff_gnn4items.loc['GIN'].precision, fmts[0])
axs[3, 1].legend(["GAT", "GEN", "GIN"])
axs[3, 1].grid(True, linestyle='--', linewidth=.57)

# recall
axs[4, 0].plot(timesteps, total_df_diff_gnn4drugs.loc['GAT'].recall, fmts[1],
               timesteps, total_df_diff_gnn4drugs.loc['GEN'].recall, fmts[3],
               timesteps, total_df_diff_gnn4drugs.loc['GIN'].recall, fmts[0])
axs[4, 0].set_ylabel('Recall')
axs[4, 0].legend(["GAT", "GEN", "GIN"])
axs[4, 0].grid(True, linestyle='--', linewidth=.57)
axs[4, 1].plot(timesteps, total_df_diff_gnn4items.loc['GAT'].recall, fmts[1],
               timesteps, total_df_diff_gnn4items.loc['GEN'].recall, fmts[3],
               timesteps, total_df_diff_gnn4items.loc['GIN'].recall, fmts[0])
axs[4, 1].legend(["GAT", "GEN", "GIN"])
axs[4, 1].grid(True, linestyle='--', linewidth=.57)

# F1
axs[5, 0].plot(timesteps, total_df_diff_gnn4drugs.loc['GAT'].F1, fmts[1],
               timesteps, total_df_diff_gnn4drugs.loc['GEN'].F1, fmts[3],
               timesteps, total_df_diff_gnn4drugs.loc['GIN'].F1, fmts[0])
axs[5, 0].set_ylabel('F1')
axs[5, 0].legend(["GAT", "GEN", "GIN"])
axs[5, 0].grid(True, linestyle='--', linewidth=.57)
axs[5, 1].plot(timesteps, total_df_diff_gnn4items.loc['GAT'].F1, fmts[1],
               timesteps, total_df_diff_gnn4items.loc['GEN'].F1, fmts[3],
               timesteps, total_df_diff_gnn4items.loc['GIN'].F1, fmts[0])
axs[5, 1].legend(["GAT", "GEN", "GIN"])
axs[5, 1].grid(True, linestyle='--', linewidth=.57)

# PRAUC
axs[6, 0].plot(timesteps, total_df_diff_gnn4drugs.loc['GAT'].PRAUC, fmts[1],
               timesteps, total_df_diff_gnn4drugs.loc['GEN'].PRAUC, fmts[3],
               timesteps, total_df_diff_gnn4drugs.loc['GIN'].PRAUC, fmts[0])
axs[6, 0].set_ylabel('PRAUC')
axs[6, 0].legend(["GAT", "GEN", "GIN"])
axs[6, 0].grid(True, linestyle='--', linewidth=.57)
axs[6, 0].set_xlabel('day')
axs[6, 1].plot(timesteps, total_df_diff_gnn4items.loc['GAT'].PRAUC, fmts[1],
               timesteps, total_df_diff_gnn4items.loc['GEN'].PRAUC, fmts[3],
               timesteps, total_df_diff_gnn4items.loc['GIN'].PRAUC, fmts[0])
axs[6, 1].legend(["GAT", "GEN", "GIN"])
axs[6, 1].grid(True, linestyle='--', linewidth=.57)
axs[6, 1].set_xlabel('day')

# fig.suptitle('Performance of different downstream recommendation tasks \n')
plt.tight_layout()
```

```python
# fmts = ('.k--', 'xm--', 'vg-.', '*r:')
fmts = ('.C7:', '.C5--', '.C2:', '.C3-')
timesteps = np.arange(1, 20+1)

fig, axs = plt.subplots(1, 2, sharex="col", sharey="row", figsize=(4*2, 3*1))

# TRUE
axs[0].plot(timesteps, total_df_diff_gnn4drugs.loc['GAT'].DDI_true, fmts[1],
            timesteps, total_df_diff_gnn4drugs.loc['GEN'].DDI_true, fmts[3],
            timesteps, total_df_diff_gnn4drugs.loc['GIN'].DDI_true, fmts[0])
axs[0].grid(True, linestyle='--', linewidth=.57)
axs[0].set_title("ground ture")
axs[0].set_ylabel('DDI')
axs[0].set_xlabel('day')
axs[0].legend(["GAT", "GEN", "GIN"])

# PRED
axs[1].plot(timesteps, total_df_diff_gnn4drugs.loc['GAT'].DDI_pred, fmts[1],
            timesteps, total_df_diff_gnn4drugs.loc['GEN'].DDI_pred, fmts[3],
            timesteps, total_df_diff_gnn4drugs.loc['GIN'].DDI_pred, fmts[0])
axs[1].set_title("model predict")
axs[1].grid(True, linestyle='--', linewidth=.57)
axs[1].set_xlabel('day')
axs[1].legend(["GAT", "GEN", "GIN"])

plt.tight_layout()
```
