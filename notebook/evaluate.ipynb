```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path_dfs = r"../results/dfs"
list_csvgz = [f for f in os.listdir(path_dfs) if f[-3:]=="csv"]
list_csvgz
```

```python
dfs = [pd.read_csv(os.path.join(path_dfs, f)) for f in list_csvgz]
dfs[0]
```

# Different GNN

```python
total_df_diff_gnn = pd.concat(dfs[:3], keys=["GAT", "GEN", "GIN"])
total_df_diff_gnn.loc['GAT']
```

```python
total_df_diff_gnn.loc['GAT'].auc.mean(), total_df_diff_gnn.loc['GEN'].auc.mean(), total_df_diff_gnn.loc['GIN'].auc.mean()
```

```python
total_df_diff_gnn.loc['GAT'].acc.mean(), total_df_diff_gnn.loc['GEN'].acc.mean(), total_df_diff_gnn.loc['GIN'].acc.mean()
```

```python
total_df_diff_gnn.loc['GAT'].acc.iloc[0], total_df_diff_gnn.loc['GEN'].acc.iloc[0], total_df_diff_gnn.loc['GIN'].acc.iloc[0]
```

```python
total_df_diff_gnn.loc['GEN'].recall_marco_avg.mean()
```

```python
total_df_diff_gnn.loc['GEN'].precision_marco_avg.mean(), total_df_diff_gnn.loc['GEN'].precision_weighted_avg.mean()
```

```python
fmts = ('.C7:', '.C5--', '.C2:', '.C3-')
timesteps = range(1, 20+1)

fig, axs = plt.subplots(3, 2, sharex="all", figsize=(4*2, 3*3))

# precision
axs[0, 0].plot(timesteps, total_df_diff_gnn.loc['GAT'].precision_marco_avg, fmts[1],
               timesteps, total_df_diff_gnn.loc['GEN'].precision_marco_avg, fmts[3],
               timesteps, total_df_diff_gnn.loc['GIN'].precision_marco_avg, fmts[0])
axs[0, 0].set_title('marco average', fontsize=13)
axs[0, 0].set_ylabel('precision', fontsize=13)
axs[0, 0].legend(["GAT", "GEN", "GIN"])
axs[0, 0].grid(True, linestyle='--', linewidth=.57)
axs[0, 1].plot(timesteps, total_df_diff_gnn.loc['GAT'].precision_weighted_avg, fmts[1],
               timesteps, total_df_diff_gnn.loc['GEN'].precision_weighted_avg, fmts[3],
               timesteps, total_df_diff_gnn.loc['GIN'].precision_weighted_avg, fmts[0])
axs[0, 1].set_title('weighted average', fontsize=13)
axs[0, 1].legend(["GAT", "GEN", "GIN"])
axs[0, 1].grid(True, linestyle='--', linewidth=.57)

# recall
axs[1, 0].plot(timesteps, total_df_diff_gnn.loc['GAT'].recall_marco_avg, fmts[1],
               timesteps, total_df_diff_gnn.loc['GEN'].recall_marco_avg, fmts[3],
               timesteps, total_df_diff_gnn.loc['GIN'].recall_marco_avg, fmts[0])
axs[1, 0].set_ylabel('recall', fontsize=13)
axs[1, 0].legend(["GAT", "GEN", "GIN"])
axs[1, 0].grid(True, linestyle='--', linewidth=.57)
axs[1, 1].plot(timesteps, total_df_diff_gnn.loc['GAT'].recall_weighted_avg, fmts[1],
               timesteps, total_df_diff_gnn.loc['GEN'].recall_weighted_avg, fmts[3],
               timesteps, total_df_diff_gnn.loc['GIN'].recall_weighted_avg, fmts[0])
axs[1, 1].legend(["GAT", "GEN", "GIN"])
axs[1, 1].grid(True, linestyle='--', linewidth=.57)

# F1
axs[2, 0].plot(timesteps, total_df_diff_gnn.loc['GAT'].f1_marco_avg, fmts[1],
               timesteps, total_df_diff_gnn.loc['GEN'].f1_marco_avg, fmts[3],
               timesteps, total_df_diff_gnn.loc['GIN'].f1_marco_avg, fmts[0])
axs[2, 0].set_ylabel('F1-score', fontsize=13)
axs[2, 0].set_xlabel('day')
axs[2, 0].legend(["GAT", "GEN", "GIN"])
axs[2, 0].grid(True, linestyle='--', linewidth=.57)
axs[2, 1].plot(timesteps, total_df_diff_gnn.loc['GAT'].f1_weighted_avg, fmts[1],
               timesteps, total_df_diff_gnn.loc['GEN'].f1_weighted_avg, fmts[3],
               timesteps, total_df_diff_gnn.loc['GIN'].f1_weighted_avg, fmts[0])
axs[2, 1].set_xlabel('day')
axs[2, 1].legend(["GAT", "GEN", "GIN"])
axs[2, 1].grid(True, linestyle='--', linewidth=.57)

# fig.suptitle('Performance of different types of GNN per day', fontsize=17)
plt.tight_layout()
```

```python
# fmts = ('.k--', 'xm--', 'vg-.', '*r:')
fmts = ('.C7:', '.C5--', '.C2:', '.C3-')
timesteps = np.arange(1, 20+1)

fig, axs = plt.subplots(1, 2, sharex="all", figsize=(4*2, 3*1))

# AUC
axs[0].plot(timesteps, total_df_diff_gnn.loc['GAT'].auc, fmts[1],
            timesteps, total_df_diff_gnn.loc['GEN'].auc, fmts[3],
            timesteps, total_df_diff_gnn.loc['GIN'].auc, fmts[0])
axs[0].grid(True, linestyle='--', linewidth=.57)
axs[0].set_ylabel('AUC', fontsize=13)
axs[0].set_xlabel('day')
axs[0].legend(["GAT", "GEN", "GIN"])

# ACC
axs[1].plot(timesteps, total_df_diff_gnn.loc['GAT'].acc, fmts[1],
            timesteps, total_df_diff_gnn.loc['GEN'].acc, fmts[3],
            timesteps, total_df_diff_gnn.loc['GIN'].acc, fmts[0])
axs[1].grid(True, linestyle='--', linewidth=.57)
axs[1].set_ylabel('ACC', fontsize=13)
axs[1].set_xlabel('day')
axs[1].legend(["GAT", "GEN", "GIN"])

plt.tight_layout()
```
